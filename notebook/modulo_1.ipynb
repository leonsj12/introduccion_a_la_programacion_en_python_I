## Grafico de arbol de navidad

El código extradido con el módulo *turtle* no funciona directamente en Google Colab, ya que turtle depende de una interfaz gráfica que no está disponible en este entorno. Sin embargo, lo adaptamos para Colab utilizando bibliotecas como matplotlib para dibujar gráficos en lugar de *turtle*.

**Ajustes realizados:**
- Reemplazamos las funciones de dibujo de *turtle* por las correspondientes en matplotlib.
- Utilizamos *patches.Circle* y *patches.Rectangle* para dibujar los círculos y cuadrados.
- Ajustamos los límites del lienzo para que sean consistentes con el espacio de trabajo de turtle.

```python
# Codigo extraido de: https://gist.github.com/jurandysoares/4380835

import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Configuración inicial del lienzo
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(-300, 300)
ax.set_ylim(-300, 300)
ax.set_aspect('equal')
ax.axis('off')  # Ocultar los ejes

# Función para dibujar un círculo
def draw_circle(x, y, color):
    circle = patches.Circle((x, y), 15, color=color, ec="black", lw=0.5)
    ax.add_patch(circle)

# Función para dibujar un cuadrado
def draw_square(x, y, color):
    square = patches.Rectangle((x - 15, y - 15), 30, 30, color=color, ec="black", lw=0.5)
    ax.add_patch(square)

# Dibujar el círculo inicial en la parte superior
draw_circle(0, 280, 'red')

k = 0
# Dibujar las filas principales
for i in range(1, 17):
    y = 280 - 30 * i
    for j in range(i - k):
        x = 30 * j
        draw_square(x, y, 'green')
        draw_square(-x, y, 'green')

    if i % 4 == 0:
        x = 30 * (j + 1)
        draw_circle(-x, y, 'red')
        draw_circle(x, y, 'red')
        k += 2

    if i % 4 == 3:
        x = 30 * (j + 1)
        draw_circle(-x, y, 'yellow')
        draw_circle(x, y, 'yellow')

# Dibujar el tronco del árbol
for i in range(17, 20):
    y = 280 - 30 * i
    for j in range(3):
        x = 30 * j
        draw_square(x, y, 'brown')
        draw_square(-x, y, 'brown')

# Mostrar el dibujo
plt.show()
```
